```python
import numpy as np
import pandas as pd
import seaborn as sns

sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('911.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
# the five top Zipcodes 
df['zip'].value_counts().head(5)
```

```python
# How many unique Titles
df['title'].nunique()
```

```python
#In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. 
#I create a new column called "Reason" that contains this string value.
df['Reason']=df['title'].apply(lambda x: x.split(':')[0])
df.head()
```

```python
#the most common Reason for a 911 call based off of this new column
df['Reason'].value_counts()
```

```python
sns.countplot(x='Reason',data=df)
```

```python
#the data type of the objects in the timeStamp column
type(df['timeStamp'].iloc[0])
```

```python
#convert the column from strings to DateTime objects
df['timeStamp']= pd.to_datetime(df['timeStamp'])
type(df['timeStamp'].iloc[0])
```

```python
# grab specific attributes from a Datetime object by calling them
time = df['timeStamp'].iloc[0]
time.hour

#create 3 new columns called Hour, Month, and Day of Week.
df['Hour']=df['timeStamp'].apply(lambda time: time.hour)
df['Month']=df['timeStamp'].apply(lambda time: time.month)
df['Day of Week']=df['timeStamp'].apply(lambda time: time.dayofweek)

# keep the Date
df['Date']=df['timeStamp'].apply(lambda t: t.date())

#Convert Day of the week from Int to Str
dmap = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day of Week'] = df['Day of Week'].map(dmap)
```

```python
df.head()
```

```python
#create a countplot of the Day of Week column with the hue based off of the Reason column.
sns.countplot(x='Day of Week',hue='Reason',data=df,palette='viridis')
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
```

```python
#create a countplot of the Day of Week column with the hue based off of the Reason column.
sns.countplot(x='Month',hue='Reason',data=df,palette='PuBuGn')
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
```
